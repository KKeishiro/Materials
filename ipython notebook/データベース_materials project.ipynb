```python
atomicdata =pd.read_csv("/media/sf_kyoyu/data/atomicdata_2015.5edition.csv").set_index("Symbol")
compoundtable=pd.read_csv("/media/sf_kyoyu/data/compound_table.csv").dropna()\
.drop(["Ecoh","BM","bandgap","Compound","type","st"],axis=1).sort("formula")
compoundtable=compoundtable[~compoundtable["formula"].duplicated(take_last=False)].set_index("formula")
```

```python
def make_average_data():
    atomicdata.ix['None',:] = 0
    average_data = []
    for i in np.array(compoundtable.index):
        a1 = compoundtable.loc[i,'a1']
        a2 = compoundtable.loc[i,'a2']
        a3 = compoundtable.loc[i,'a3']
        n1 = compoundtable.loc[i,'n1']
        n2 = compoundtable.loc[i,'n2']
        n3 = compoundtable.loc[i,'n3']
        row_data = [((atomicdata.loc[a1,j]*n1)+(atomicdata.loc[a2,j]*n2)+(atomicdata.loc[a3,j]*n3))/(n1+n2+n3) \
                    for j in np.array(atomicdata.columns)]
        average_data.append(row_data)
    frame = pd.DataFrame(average_data, index = compoundtable.index, columns = atomicdata.columns)
    frame.to_csv("/media/sf_kyoyu/data/average_database_limited.csv")
```

```python
def get_matpro_data():
    a = MPRester("m390ovHWT4KfcLu2")
    data_list = []
    for i in np.array(A.index):
        entries = a.get_entries(i,property_data=["energy","energy_per_atom","volume"])
        for entry in entries:
            name = entry.name
            e = entry.data["energy"]
            epa = entry.data["energy_per_atom"]
            volume = entry.data["volume"]
            data = [name,e,epa,volume]
            data_list.append(data)
    frame = pd.DataFrame(data_list, columns = ["name","energy","energy_per_atom","volume"]).set_index("name")
    frame.to_csv("/media/sf_kyoyu/data/matpro_data.csv")
```

```python
# 平均ー逆数データ
C_inverse = 1/C
# C_inverse = C_inverse.replace('inf',0)

# 平均ー二乗データ
C_square = C*C

# 平均ー対数データ
C_log = np.log(C)
# C_log = C_log.replace('-inf',0).fillna(0)
```

```python
# 標準化
# -----------------説明変数の標準化
# 平均そのままデータ
scaled_list_x = []
for i in np.array(C.columns):
    sc = StandardScaler()
    sc.fit(C.loc[:,i])
    array_scaled_x = sc.transform(C.loc[:,i])
    scaled_list_x.append(array_scaled_x)
df_x = pd.DataFrame(scaled_list_x).T
df_x.index = np.array(C.index)
df_x.columns = np.array(C.columns)

# 平均ー逆数データの標準化
scaled_list_inverse = []
for i in np.array(C_inverse.columns):
    sc = StandardScaler()
    sc.fit(C_inverse.loc[:,i])
    array_scaled_inverse = sc.transform(C_inverse.loc[:,i])
    scaled_list_inverse.append(array_scaled_inverse)
df_x_inverse = pd.DataFrame(scaled_list_inverse).T
df_x_inverse.index = np.array(C_inverse.index)
df_x_inverse.columns = np.array(C_inverse.columns)

# 平均ー二乗データの標準化
scaled_list_square = []
for i in np.array(C_square.columns):
    sc = StandardScaler()
    sc.fit(C_square.loc[:,i])
    array_scaled_square = sc.transform(C_square.loc[:,i])
    scaled_list_square.append(array_scaled_square)
df_x_square = pd.DataFrame(scaled_list_square).T
df_x_square.index = np.array(C_square.index)
df_x_square.columns = np.array(C_square.columns)

# 平均ー対数データの標準化
scaled_list_log = []
for i in np.array(C_log.columns):
    sc = StandardScaler()
    sc.fit(C_log.loc[:,i])
    array_scaled_log = sc.transform(C_log.loc[:,i])
    scaled_list_log.append(array_scaled_log)
df_x_log = pd.DataFrame(scaled_list_log).T
df_x_log.index = np.array(C_log.index)
df_x_log.columns = np.array(C_log.columns)

# 平均ー全データ連結
df_x_all = pd.concat([df_x,df_x_inverse,df_x_square,df_x_log], axis=1)
```

```python
# 主成分分析
pca = PCA(n_components = 16, whiten = False)
pca.fit(df_x_all)
df_pca_x_all = pca.transform(df_x_all)
E = pca.explained_variance_ratio_
# pca_components = pd.DataFrame(pca.components_ ,index = ["第一主成分","第二主成分"], columns = np.array(df_x.columns))
df_pca_x_all = pd.DataFrame(df_pca_x_all, columns = ["第一主成分","第二主成分","第三主成分","第四主成分","第五主成分","第六主成分",\
                                         "第七主成分","第八主成分","第九主成分","第十主成分","第十一主成分","第十二主成分"\
                                            ,"第十三主成分","第十四主成分","第十五主成分","第十六主成分"\
                                            ], index = df_x_all.index)
# print pca.components_
print np.cumsum(E)
```
